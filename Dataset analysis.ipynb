```python
import pandas as pd

# Load dataset

file_path = 'Authentic-48K.csv'
df = pd.read_csv(file_path, encoding='latin1', low_memory=False)

print("NEWSPAPER ANALYSIS")

# 1. newspapers (domains) in this file
unique_newspapers = df['domain'].nunique()
print(f"Number of unique newspapers: {unique_newspapers}")

# 2. Top 10 most frequent newspapers
print("\nTop 10 Newspapers by Article Count:")
print(df['domain'].value_counts().head(10))


print(" CATEGORY DISTRIBUTION ")

# 3.  level of distribution across categories
# This shows you which topics (Politics, Sports, etc.) dominate the data
category_dist = df['category'].value_counts()
print(category_dist)

# 4. Percentage distribution (The "Level")

print("\nPercentage Distribution:")

print(df['category'].value_counts(normalize=True) * 100)

#5.
print(f"TOTAL NUMBER OF NEWSPAPERS: {df['domain'].nunique()}")

#6.  news that is spread across different topics
print(df['category'].value_counts(normalize=True) * 100)
```
